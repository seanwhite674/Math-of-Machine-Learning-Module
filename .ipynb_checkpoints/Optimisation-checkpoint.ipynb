# Unconstrained optimisation

## Load useful packages

Before we start, let's load [numpy](https://numpy.org) and [matplotlib](https://matplotlib.org) (plotting).

```python
import numpy as np
import numpy.linalg as npl 
import matplotlib.pyplot as plt
```

## Creating a quadratic optimisation problem
Let's create a random quadratic optimisation problem with condition number 100. 

```python
# Set the random number generator seed so that the notebook is exactly reproducible
np.random.seed(42)

# Create a digonal matrix with two entries
D = np.diag([1,3])

# Create random orthogonal matrix
H = np.random.randn(2, 2)
Q, R = npl.qr(H)

# Create A and b
A = Q.T@D@Q
xstar = np.array([1,1])
b = A@xstar
```

```python
A=np.array([[7,-4],[-4,7]])
xstar = np.array([1,1])
b = A@xstar
```

```python
np.linalg.eig(A)
```

## Gradient descent

The quadratic optimisation problem 
$$ \min_x \; f(x)=\frac{1}{2}x^T A x - x^T b$$
can be solved via gradient descent method.
$$ x^{(k+1)} = x^{(k)} - \alpha_k g^{(k)}$$
where $g^{(k)}=\nabla f(x^{(k)})=Ax^{(k)}-b$ and an optimal $\alpha_k$ can be determined at each step as $\alpha_k = \frac{\Vert g^{(k)} \Vert^2}{(g^{(k)})^T A g^{(k)}}$.

Now we create a function for the gradient descent method for a generic *quadratic programming* problem.

```python
def SD(A, b, x, tol, maxit):
    xvals = [x]
    func = [0.5*x@A@x]

    rk = A@x-b
    iter = 0
    converged = False

    while not converged and iter<maxit:
        alphak = (rk@rk)/(rk@A@rk)
        x = x - alphak*rk
        rk = A@x-b

        converged = npl.norm(rk)<tol
        xvals.append(x)
        func.append(0.5*x@A@x - x@b)

        iter += 1
    
    return x, np.array(xvals), np.array(func)
```

Run the gradient descent for our specific problem with an initial guess of $x_0 = (5, 5)^T$.

```python
x0     = np.array([4,7])+ xstar
tol   = 1.e-9
maxit = 1000
x_sd, xvals_sd, funvals_sd = SD(A, b, x0, tol, maxit)
```

Plot the solution

```python
xx = np.linspace(np.min(xvals_sd[:,0])-1,np.max(xvals_sd[:,0])+1,500)
yy = np.linspace(np.min(xvals_sd[:,1])-1,np.max(xvals_sd[:,1])+1,500)
Xgrid,Ygrid = np.meshgrid(xx,yy)

Z = np.zeros(Xgrid.shape)

for i in range(Xgrid.shape[0]):
    for j in range (Xgrid.shape[1]):
        v = np.array([Xgrid[i,j],Ygrid[i,j]])
        Z[i,j] = 0.5*v@A@v-v@b
```

```python
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(2, 1, 1)
plt.contourf(Xgrid, Ygrid, Z, levels=25)
plt.colorbar()
plt.plot(xvals_sd[:,0], xvals_sd[:,1], lw = 1, c = "orange")

ax = fig.add_subplot(2, 1, 2, projection='3d')
n = 150
ax.contour3D(Xgrid,Ygrid,Z, n, alpha = .5)
ax.view_init(elev=30, azim=-60)
ax.plot(xvals_sd[:,0], xvals_sd[:,1], funvals_sd, c = "orange", lw = 1)
plt.show()
```

## Gradient descent with momentum

We have also introduced the gradient descent method with momentum.
$$ x^{(k+1)} = x^{(k)} - \alpha z^{(k)}$$
$$ z^{(k+1)} = \nabla f(x^{(k+1)}) + \beta z^{(k)}$$
where an optimal choice for $\alpha$ and $\beta$ can be determined as
$$\alpha = \left(\frac{2}{\sqrt{\lambda_{max}(A)}+\sqrt{\lambda_{min}(A)}}\right)^2,
\qquad \beta = \left(\frac{\sqrt{\lambda_{max}(A)}-\sqrt{\lambda_{min}(A)}}{\sqrt{\lambda_{max}(A)}+\sqrt{\lambda_{min}(A)}}\right)^2$$

Let's now implement the method:

```python
def grad_momentum(A, b, x, alpha, beta, tol, maxit):
    xvals = [x]
    func = [0.5*x@A@x - x@b]
    
    grad = A@x-b
    zk = grad

    iter = 0
    converged = False
    while not converged and iter<maxit:
        x = x - alpha*zk
        grad = A@x-b
        zk = grad + beta*zk

        converged = npl.norm(grad)<tol
        xvals.append(x)
        func.append(0.5*x@A@x - x@b)

        iter += 1
    
    return x, np.array(xvals), np.array(func)
```

Try it with our example problem:

```python
lam_max, lam_min = np.linalg.eigvals(A)
alpha = (2/(np.sqrt(lam_max)+np.sqrt(lam_min)))**2
beta = ((np.sqrt(lam_max)-np.sqrt(lam_min))/(np.sqrt(lam_max)+np.sqrt(lam_min)))**2
```

```python
tol=1.e-9
maxit = 1000
```

```python
x_gm, xvals_gm, funvals_gm = grad_momentum(A, b, x0, alpha, beta, tol, maxit)
```

```python
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(2, 1, 1)
plt.contourf(Xgrid, Ygrid, Z, levels=25)
plt.colorbar()
plt.plot(xvals_sd[:,0],xvals_sd[:,1], lw = 1.5, c = "orange")
plt.plot(xvals_gm[:,0],xvals_gm[:,1], lw = 1.5, c = "red")

ax = fig.add_subplot(2, 1, 2, projection='3d')
n = 150
ax.contour3D(Xgrid,Ygrid,Z, n, alpha = .5)
ax.plot(xvals_sd[:,0],xvals_sd[:,1],funvals_sd, c = "orange", lw = 1.5)
ax.plot(xvals_gm[:,0],xvals_gm[:,1],funvals_gm, c = "red", lw = 1.5)
plt.show()
```

Let's look at how the error decreaseds at each iteration. It is faster with momentum:

```python
fopt = 0.5*(xstar@A@xstar)-xstar@b
err_sd = funvals_sd-fopt
err_gm = funvals_gm-fopt

plt.semilogy(err_sd,'b-',label='Steepest Descent')
plt.semilogy(err_gm,'r:',label='Momentum Gradient')
plt.legend()
plt.show()
```

## General nonlinear optimisation problems

Consider the Rosenbrock function,$$f(x,y)=(x−1)^2+ b(y−x^2)^2$$
for the choice $b=10$. The gradient of $f(x,y)$ is
$$ \nabla f(x,y) = \begin{bmatrix} 2(x-1) - 4b(y - x^2)x \\
                         2b(y-x^2)\end{bmatrix}.$$
Now let's define these in Python:

```python
b = 10;
f = lambda x: (x[0]-1)**2 + b*(x[1]-x[0]**2)**2;
df = lambda x: np.array([2*(x[0]-1) - 4*b*(x[1] - x[0]**2)*x[0],
                         2*b*(x[1]-x[0]**2)])
```

We will now define functions for gradient descent with and without momentum to solve this nonlinear optimisation problem:

```python
def SD_nonlin(f, gradf, x, alpha, tol, maxit):
    xvals = [x]
    func = [f(x)]
    grad = gradf(x)

    iter = 0
    converged = False
    while not converged and iter<maxit:
        x = x - alpha*grad
        grad = gradf(x)

        converged = npl.norm(grad)<tol
        xvals.append(x)
        func.append(f(x))
        iter += 1
    return x, np.array(xvals), np.array(func)
```

```python
def grad_momentum_nonlin(f,gradf,x,alpha,beta,tol,maxit):
    xvals = [x]
    func = [f(x)]
    
    grad = gradf(x)
    zk = grad

    iter = 0
    converged = False
    while not converged and iter<maxit:
        x = x - alpha*zk
        grad = gradf(x)
        zk = grad + beta*zk

        converged = npl.norm(grad)<tol
        xvals.append(x)
        func.append(f(x))
        iter += 1
    
    return x, np.array(xvals), np.array(func)
```

Let's now find a minimiser of the Rosenbrock function using both methods.

```python
x0 = np.array([-1.4, 1.1])
tol=1.e-9
alpha = 0.02
alphaM = 0.02
beta = 0.8
maxit = 1000
x_sd, xvec_sd, funvec_sd = SD_nonlin(f,df,x0,alpha,tol,maxit)
x_gm, xvec_gm, funvec_gm = grad_momentum_nonlin(f,df,x0,alphaM,beta,tol,maxit)
```

```python
xx = np.arange(-2, 2, 0.15)
yy = np.arange(-1, 3, 0.15)
Xgrid,Ygrid = np.meshgrid(xx,yy)
Z = np.zeros(Xgrid.shape)

for i in range(Xgrid.shape[0]):
    for j in range (Xgrid.shape[1]):
        v = np.array([Xgrid[i,j],Ygrid[i,j]])
        Z[i,j] = f(v)
```

```python
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(2, 1, 1)
plt.contourf(Xgrid,Ygrid,Z, levels=25)
plt.colorbar()
plt.plot(xvec_sd[:,0],xvec_sd[:,1], lw = 3, c = "orange")
plt.plot(xvec_gm[:,0],xvec_gm[:,1], lw = 3, c = "red",linestyle='-.')
plt.plot(1,1,marker='o',markersize=5, color ='g')

ax = fig.add_subplot(2, 1, 2, projection='3d')
n = 150
ax.contour3D(Xgrid,Ygrid,Z, n, alpha = .5)
ax.plot(xvec_sd[:,0],xvec_sd[:,1],funvec_sd, c = "orange", lw = 3)
ax.plot(xvec_gm[:,0],xvec_gm[:,1],funvec_gm, c = "red", lw = 3,linestyle='-.')
plt.plot(1,1,0,marker='o',markersize=5, color ='g')

plt.show()
```

```python
plt.figure(1)
plt.semilogy(funvec_sd,'b-',label='Steepest Descent')
plt.semilogy(funvec_gm,'r:',label='Momentum Gradient')
plt.legend()
plt.show()
```

## Stochastic gradient descent

Let's create a random linear regression problem. 

```python
np.random.seed(42)
x = np.pi - 2*np.pi*np.random.rand(500)
y = 3*np.sin(x) + 2*np.cos(5*x) + 0.5*np.random.randn(500)
fig = plt.figure(figsize=(8,6))
plt.scatter(x, y)
plt.show()
```

We now build a linear regression problem to fit the data to a model of the form
$$ y(x;a,b)=a \sin(x) + b \cos(5x)$$

This can be done by considering an (overdetermined) linear system of equations $ A z = y$ where
$$ A = \begin{pmatrix} \sin(x_1) & \cos(5x_1) \\ \vdots & \vdots \\ \sin(x_n) & \cos(5x_n) \end{pmatrix}, \quad y = \begin{pmatrix} y_1 \\ \vdots \\ y_n \end{pmatrix}, \quad z = \begin{pmatrix} a \\ b \end{pmatrix}$$

We look at the value of $z$ which minimises the scaled square L2-norm $f(z) = \frac{1}{2n}||Az - y||^2$. First, we construct the matrix $A$:

```python
A = np.array([np.sin(x),np.cos(5*x)]).T
A.shape
```

### Solve the optimisation problem using stochastic gradient descent
The objective function can be rewritten as
$$ f(z) = \frac{1}{n}\sum_{i=1}^{n} \frac{1}{2}(A_i z-y_i)^2 =: \frac{1}{n}\sum_{i=1}^{n}f_i(z),$$
where $A_i$ is the i-th row of $A$.
We have seen how the optimisation problem can be solved via stochastic gradient descent method.
$$ z^{(k+1)} = z^{(k)} - \alpha g_k(z^{(k)})$$
where $g_k(z^{(k)})$ is obtained by averaging the gradient of a subset of functions, i.e.
$$ g_k(z^{(k)}) = \frac{1}{|\mathcal{B}_k|} \sum_{i\in\mathcal{B}_k}\nabla f_i(z^{(k)})$$


```python
def SGD_4_LS(A,b,x,alpha,tol,maxit): 
    batch_size = 5
    xvec = [x]
    n = A.shape[0]
    res = A@x-b
    gradx = (A.T@(res))/n
    funcx = [0.5*(npl.norm(res)**2)/n]
    iter = 0
    converged = False
    while not converged and iter < maxit:
        i = np.random.choice(n,size=batch_size,replace=False)
        a = A[i,:]
        gi = (a.T@(a@x-b[i]))/batch_size
        x = x - alpha*gi
        xvec.append(x)
        res = A@x-b
        gradx = (A.T@(res))/n
        funcx.append(0.5*(npl.norm(res)**2)/n)
        iter += 1
    
    return x, np.array(xvec), np.array(funcx)    
```

```python
x0 = np.array([1,2])
tol=1.e-8
alpha = 0.01
maxit = 10000
np.random.seed(42)
z_sgd, zvec_sgd, funvec_sgd = SGD_4_LS(A, y, x0, alpha, tol, maxit)
```

```python
plt.semilogy(funvec_sgd)
plt.show()
```

```python
xx = np.linspace(0.5,3.5,200)
yy = np.linspace(1,3,200)
Xgrid,Ygrid = np.meshgrid(xx,yy)
Z = np.zeros(Xgrid.shape)

for i in range(Xgrid.shape[0]):
    for j in range (Xgrid.shape[1]):
        Z[i,j] = 0.5*(npl.norm(A@[Xgrid[i,j],Ygrid[i,j]]-y)**2)/500
```

```python
fig = plt.figure(figsize=(10,15))

ax = fig.add_subplot(2, 1, 1)
plt.contourf(Xgrid,Ygrid,Z, levels=25)
plt.colorbar()
plt.plot(zvec_sgd[:,0],zvec_sgd[:,1],'+', c = "orange")
```

```python
xmod = np.sort(x,axis=None)
ymod = z_sgd[0]*np.sin(xmod)+z_sgd[1]*np.cos(5*xmod)
fig = plt.figure(figsize=(8,6))
plt.scatter(x, y,label='Data')
plt.plot(xmod,ymod,color='m',linewidth=3,linestyle='-.',label='Fitted model')
plt.title("Fitted linear model with gradient descent")
plt.show()
```

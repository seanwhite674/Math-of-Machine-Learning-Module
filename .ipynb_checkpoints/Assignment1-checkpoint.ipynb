# Q 4 b)

```python
import numpy as np
```

```python
A = np.array([[2,1,3],[1,9,1],[4,6,5]])
n = A.shape[0]
```

```python
U = np.array(A,dtype = float)
L = np.identity(n)

for j in range(0,n):
    for i in range(j+1,n):
        L[i,j] = (U[i,j]/U[j,j])
        U[i,:] = U[i,:] - L[i,j]*U[j,:]
```

```python
L@U
```

```python
A = np.array([[2,1,3],[1,9,1],[4,6,5]])
n = A.shape[0]
```

### Logic need n elemtary matrices E

```python
def E2(i,j,c):
    e2 = np.identity(n)
    e2[i, j] = c
    return e2   
```

```python
E2(1,0,-1/2)
```

```python
E2(2,0,-2)
```

```python
E2(2,1,-8/17)
```

# $$E_{21}E_{31}E_{32}A=U$$
# $$ A=(E_{32})^{-1}(E_{31})^{-1}(E_{21})^{-1}U$$
# $$ L= (E_{32})^{-1}(E_{31})^{-1}(E_{21})^{-1}$$
### Inverse of identity matrix with one other entry. Just change sign of other entry

```python
L=E2(1,0,1/2)@E2(2,0,2)@E2(2,1,8/17)
L
```

```python
U = E2(2,1,-8/17)@E2(2,0,-2)@E2(1,0,-1/2)@A
U
```

```python
L@U
```

```python
A = np.array([[2,1,3],[1,9,1],[4,6,5]])
n = 3
```

```python
u1 = a1
normu1 = norm(u1)
normu1
```

```python
q1 = u1/normu1
q1
```

```python
u2 = a2 - (a2@u1)/(u1@u1)*u1
u2
```

```python
q2 = u2/norm(u2)
q2
```

```python
Q=np.transpose(np.array([q1,q2,q3]))
Q
```

```python
A = np.array([[2,1,3],[1,9,1],[4,6,5]])
(a1, a2, a3) = np.transpose(A)
n = 3
```

```python
u1 = a1 - norm(a1)*np.array([1,0,0])
v1 = u1/norm(u1)
H1 = np.identity(3) - 2*np.outer(v1,v1)
H1
```

```python
A1=H1@A
A1
```

```python
a2 = A1[1:,1]
#u2 = a2 - (-np.sign(a2[0]))*norm(a2)*np.array([1,0])
u2 = a2 - norm(a2)*np.array([1,0])
v2 = u2/norm(u2)
H2 = np.identity(3)
H2[1:,1:] -= 2*np.outer(v2,v2)
H2
```

```python
A2=H2@A1
A2
```

```python
a3 = A2[2:,2]
#u3 = a3 - (-np.sign(a3[0]))*norm(a3)*np.array([1])
u3 = a3 - norm(a3)*np.array([1])
v3 = u3/norm(u3)
H3 = np.identity(3)
H3[2:,2:] -= 2*np.outer(v3,v3)
H3
```

```python
A3 = H3@A2
A3
```

```python
R=A3
R
```

```python
Q = H1.T@H2.T@H3.T
Q
```

```python
Acheck = Q@R
Acheck
```

# Q6 a)
## Showing My Workings

```python
from numpy.linalg import eig
A = np.array([[2,1,3],[1,9,1],[4,6,5]])
n=A.shape[0]
```

### First find the eigenvalues and vectors of $A^{T}A$ and the columns of V are these eigenvectors

```python
ATA=A.T@A
ATA
```

```python
lambda3 = eig(ATA)[0][0]
print("lambda3=%0.3f"%lambda3)
lambda2 = eig(ATA)[0][2]
print("lambda2=%0.3f"%lambda2)
lambda1 = eig(ATA)[0][1]
print("lambda1=%0.3f"%lambda1)
```

```python
eigvect3 = eig(ATA)[1][:,0]
eigvect3 = eigvect3/norm(eigvect3)
print("eigvector3=%a"%eigvect3)
eigvect2 = eig(ATA)[1][:,2]
eigvect2 = eigvect2/norm(eigvect2)
print("eigvector2=%a"%eigvect2)
eigvect1 = eig(ATA)[1][:,1]
eigvect1 = eigvect1/norm(eigvect1)
print("eigvector1=%a"%eigvect1)
```

```python
V = np.transpose(np.array([eigvect3,eigvect2,eigvect1]))
V
```

```python
sigma = np.diag(np.sqrt(np.sort(eig(ATA)[0])[::-1]))
sigma
```

####  Since we have $$Av_i=\sigma_iu_i$$ and so $$u_i=\frac{1}{\sigma_i}Av_i$$

```python
u1 = (1/sigma[0,0])*A@V[:,0]
u2 = (1/sigma[1,1])*A@V[:,1]
u3 = (1/sigma[2,2])*A@V[:,2]
U = np.transpose(np.array([u1,u2,u3]))
U
```

```python
Acheck = U@sigma@V.T
Acheck
```

## Can also get the SVD using the eigenvectors of $AA^{T}$ which are the columns of the matrix U

```python
AAT=A@A.T
AAT
```

```python
eigval1 = eig(AAT)[0][0]
print("Eigenvalue 1 = %0.3f"%eigval1)
eigval2 = eig(AAT)[0][2]
print("Eigenvalue 2 = %0.3f"%eigval2)
eigval3 = eig(AAT)[0][1]
print("Eigenvalue 3 = %0.3f"%eigval3)
```

```python
eigvector1 = eig(AAT)[1][:,0] 
print("Eigenvector 1 = %a"%eigvector1)
eigvector2 = eig(AAT)[1][:,2] 
print("Eigenvector 2 = %a"%eigvector2)
eigvector3 = eig(AAT)[1][:,1] 
print("Eigenvector 3 = %a"%eigvector3)
```

```python
U=np.transpose(np.array([eigvector1,eigvector2,eigvector3]))
U
```

####  Similarly I commented out the code below but it would also work to find the SVD. Since we have $$A^Tu_i=\sigma_iv_i$$ and so $$v_i=\frac{1}{\sigma_i}A^Tu_i$$

```python
v1 = 1/sigma[0,0]*A.T@U[:,0]
v2 = 1/sigma[1,1]*A.T@U[:,1] 
v3 = 1/sigma[2,2]*A.T@U[:,2]
V = np.transpose(np.array([v1,v2,v3]))
V
```

```python
Acheck = U@sigma@V.T
Acheck
```
